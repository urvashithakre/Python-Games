```python
import random

def random_func():
    n = random.randint(1,3)
    return n

def computer_choice(n):
    if n == 1:
        computerChoice = 'Rock'
    elif n == 2:
        computerChoice = 'Paper'
    else:
        computerChoice = 'Scissor'
    return computerChoice

def player():
    ch = int(input("1.Rock 2.Paper 3.Scissor\nChoose an option:"))
    if ch == 1:
        ch = 'Rock'
    elif ch == 2:
        ch = 'Paper'
    else:
        ch = 'Scissor'
    return ch
 

def game():
    ch_point = 0
    n_point = 0
    
    while True:
        n = random_func()
        computerChoice = computer_choice(n)
        ch = player()
        print("You chose:",ch)
        print("the computer chose", computerChoice)
               
        if ch == 'Rock' and computerChoice == 'Scissor':
            print("Rock smashed the scissor")
            ch_point+=1
        elif ch== 'Scissor' and computerChoice == 'Rock':
            print("Rock smashed the scissor")
            n_point+=1
        elif ch == 'Rock' and computerChoice == 'Paper':
            print("Paper covers rock")
            n_point+=1
        elif ch == 'Paper' and computerChoice == 'Rock':
            print("Paper covers rock")
            ch_point+=1
        elif ch == 'Scissor' and computerChoice == 'Paper':
            print("Scissor cut the paper")
            ch_point+=1
        elif ch == 'Paper' and computerChoice == 'Scissor':
            print("Scissor cut the paper")
            n_point+=1
        else:
            if ch == computerChoice:
                print("Both entered the same choice")
                ch_point+=1
                n_point+=1
        print("Your score =" , ch_point )
        print("Computer's score =" , n_point )
        print()
        play_again = input("Play again? (y/n):")        
        if play_again!= 'y':           
            if ch_point>n_point:
                print("You are the winner!")
            elif n_point>ch_point:
                print("Computer is the winner!")
            else:
                print("Its a tie!")   
            break

game() 
```
